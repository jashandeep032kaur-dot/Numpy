```python
id=np.arange(0,11,1)
patient_labels=np.arange(1000,1000+len(id),1)
age=patient[0 :11,1].flatten()
bp=patient[ 0:11,2].flatten()
heart=patient[0:11,3].flatten()
btemp=patient[0:11,4].flatten()
o2=patient[0:11,5].flatten()
barwidth=0.10
num_categories=5
group_offset=(num_categories-1)*barwidth/2
br1=id-group_offset
br2=id-group_offset+barwidth
br3=id-group_offset+(2*barwidth)
br4=id-group_offset+(3*barwidth)
br5=id-group_offset+(4*barwidth)
plt.bar(br1,age,color='red',width=barwidth,label='Age')
plt.bar(br2,bp,color='pink',width=barwidth,label='BP')
plt.bar(br3,heart,color='brown',width=barwidth,label='heart rate')
plt.bar(br4,btemp,color='yellow',width=barwidth,label='bodytemp')
plt.bar(br5,o2,color='green',width=barwidth,label='o2 level')
fontT={'family':'monospace', 'size':22 , 'color':'green'}
fontX={'family':'serif','size':18 , 'color': 'hotpink'}
fontY={'family':'serif','size':18,'color':'darkred'}
plt.title('patient data',fontdict=fontT)
plt.ylabel('age and bp',fontY)
plt.xlabel(' patient ID ',fontX)
plt.xticks(id,patient_labels)
plt.legend()
plt.tight_layout()
plt.show()
```
